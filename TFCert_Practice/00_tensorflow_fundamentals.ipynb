### Creating random tensors

Random tensors are tensors of some arbitrary size and random values

```python
# Create two random (but the same) tensors
random_1 = tf.random.Generator.from_seed(42) # setting seed
random_1 = random_1.normal(shape=(3,2))
random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape=(3,2))

# Are they equal?

random_1, random_2, random_1 == random_2
```

```python
random_3 = tf.random.uniform(shape=(3,2))
random_3
```

### Shuffle the order of elements in a tensor

```python
# Shuffle a tensor (valuable when you want to shuffle your data; so order doesn't effect learning)
not_shuffled = tf.constant([[10,7],
                            [3,4],
                            [2,5]])
not_shuffled.ndim
```

### Getting information from Tensors

* Shape
* Rank
* Axis or dimension
* Size

```python
# Create a rank 4 tensor (4 dimensions) # read shape from right to left
rank_4_tensor = tf.zeros(shape=[2,3,4,5])
rank_4_tensor
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
# Get various attributes of our tensor
print("Datatype of every element:", rank_4_tensor.dtype)
print("Number of dimensions (rank):", rank_4_tensor.ndim)
print("Shape of tensor:", rank_4_tensor.shape)
print("Elements along 0 axis:", rank_4_tensor.shape[0]) # number of matrix groupings
print("Elements along the last axis:", rank_4_tensor.shape[-1]) # number of columns
print("Toal number of elements in our tensor:", tf.size(rank_4_tensor).numpy()) # converts to numpy so that size isn't outputted as alot of info in an array
```

### Indexing Tensors
Tensors can be indexed just like Python lists

```python
# Get the first 2 elements of each dimension
rank_4_tensor[:2,:2,:2,:2]
```

```python
# Get first element from each dimension from each index except the final one
rank_4_tensor[:1,:1,:1,:]
```

```python
# Create a rank 2 tensor (2 dimensions)
rank_2_tensor = tf.constant([[10,7],
                             [3,4]])
rank_2_tensor.ndim, rank_2_tensor.shape
```

```python
tf.size(rank_2_tensor).numpy()
```

```python
# Get last row
rank_2_tensor[1,:]
```

```python
# Add in extra dimension to our rank 2 tensor
rank_3_tensor = rank_2_tensor[...,tf.newaxis] # ... means every axis before newaxis
rank_3_tensor
```

```python
# Alternative to tf.newaxis
tf.expand_dims(rank_2_tensor, axis=-1) # -1 means last axis
```

### Manipulating tensors (tensor operations)
**Basic Operations**
`+, -, *, /`

```python
# You can add values to a tensor using the addition operator
tensor = tf.constant([[10,7],[3,4]])
tensor + 10
```

```python
# Original tensor is unchanged
tensor
```

```python
# Multiplication
tensor * 10 # elementwise
```

```python
# Subtraction
tensor - 10
```

```python
# We can use the tensorflow built-in function
tf.multiply(tensor,10) # use tf operations instead of * or +; its been optimized
```

**Matrix Multiplication** \
One of the most common tf operations e.g. dot product between row of one matrix and column of another matrix \
Two Rules:
1. Inner dimensions must match
2. Resulting matrix has shape of outer dimensions

```python
# Matrix multiplication in tensorflow
print(tensor)
tf.linalg.matmul(tensor,tensor) # can skip linalg and just do tf.matmul
```

```python
# or you can use @
tensor @ tensor
```

```python
# Create a tensor of shape = (3,2)
X = tf.constant([[1,2],
                 [3,4],
                 [5,6]])
# Create another (3,2) tensor
Y = tf.constant([[7,8],
                 [9,10],
                 [11,12]])
X,Y
```

```python
# Try to matrix multiply tensors of same shape --> will get an error; use transpose
tf.matmul(tf.transpose(X),Y)
```

**The dot product** \
Matrix multiplication is also referred to as the dot product\
You can perform it using:
* `tf.matmul()`
* `tf.tensordot()`

```python
X,Y
```

```python
# Perform the dot product on X and Y
tf.tensordot(tf.transpose(X),Y,axes=1)
```

```python
tf.matmul(X,tf.transpose(Y))
```

```python
tf.matmul(X, tf.reshape(Y,shape=(2,3)))
```

```python
# Check the values of Y, rehape Y, and transposed Y
# Note the difference between reshape and transpose
print("Normal Y:")
print(Y,"\n")
print("Y reshaped to (2,3):")
print(tf.reshape(Y,(2,3)),"\n")
print("Y transposed:")
print(tf.transpose(Y))
```

## Aggregating tensors
Aggregating tensors = combining many individual units into a group of many units

```python
# Get the absolute values
D = tf.constant([-7,-10])
D
```

```python
# Get the absolute values
tf.abs(D)
```

Let's go through the following forms of aggregation:
* Get the minimum
* Get the maximum
* Get the mean
* Get the sum

```python
# Create a random tensor with values between 0 and 100 of size 50
E = tf.constant(np.random.randint(0,100,50))
E
```

```python
tf.size(E).numpy(), E.shape, E.ndim
```

```python
# Find the minimum NOTE: aggregation methods typically begin with reduce_
tf.reduce_min(E)
```

```python
tf.reduce_max(E)
```

```python
tf.reduce_mean(E)
```

```python
tf.reduce_sum(E)
```

```python
# For std and variance, the dtype must be floats
tf.math.reduce_std(tf.cast(E,dtype=tf.float32))
```

```python
tf.math.reduce_variance(tf.cast(E,dtype=tf.float32))
```

```python
import tensorflow_probability as tfp
tfp.stats.variance(E)
```

```python
tfp.stats.stddev(tf.cast(E,dtype=tf.float32))
```
